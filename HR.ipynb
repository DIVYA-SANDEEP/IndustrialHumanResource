```python
# Calculate TotalWorkers
df['TotalWorkers'] = (
    df['MainWorkersTotalPersons'] +
    df['MarginalWorkersTotalPersons'] +
    df['MainWorkersRuralPersons'] +
    df['MainWorkersUrbanPersons'] +
    df['MarginalWorkersRuralPersons'] +
    df['MarginalWorkersUrbanPersons']
)

# Calculate MaleFemaleRatio with safe handling for division by zero
total_males = (
    df['MainWorkersTotalMales'] +
    df['MainWorkersRuralMales'] +
    df['MainWorkersUrbanMales'] +
    df['MarginalWorkersTotalMales'] +
    df['MarginalWorkersRuralMales'] +
    df['MarginalWorkersUrbanMales']
)

total_females = (
    df['MainWorkersTotalFemales'] +
    df['MainWorkersRuralFemales'] +
    df['MainWorkersUrbanFemales'] +
    df['MarginalWorkersTotalFemales'] +
    df['MarginalWorkersRuralFemales'] +
    df['MarginalWorkersUrbanFemales']
)

df['MaleFemaleRatio'] = total_males / total_females.replace(0, 1)
```

```python
def find_outliers(df, columns):
    outliers = {}
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Clip values within the bounds
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
        
        outliers[column] = (lower_bound, upper_bound)
    return outliers

# Example usage:
outliers = find_outliers(df, ['MaleFemaleRatio'])
print(outliers)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
df_new['NICName'] = df_new['NICName'].astype(str)
```

```python
def preprocess_activity(activity):
    # Convert to lowercase
    activity = activity.lower()
    # Remove any leading/trailing whitespaces
    activity = activity.strip()
    # Remove commas
    activity = activity.replace(",", "")
    return activity

df_new["NICName"]=df_new["NICName"].apply(preprocess_activity)
```

```python
df_new["NICName"].unique()
```

```python
def label_activity(activity):
    # Label agriculture-related activities
    if "agriculture" in activity or "farming" in activity or "plant" in activity or "Animal" in activity or "animal" in activity or "crops" in activity or "rice" in activity:
        return "Agriculture"
    # Label manufacturing-related activities
    elif "manufacture" in activity or "production" in activity:
        return "Manufacturing"
    # Label mining-related activities
    elif "mining" in activity:
        return "Mining"
    # Label construction-related activities
    elif "construction" in activity:
        return "Construction"
    # Label trade-related activities
    elif "trade" in activity or "wholesale" in activity or "retail" in activity:
        return "Trade"
    # Label transportation-related activities
    elif "transport" in activity or "logistics" in activity:
        return "Transportation"
    # Label public administration-related activities
    elif "public administration" in activity or "defence" in activity:
        return "Public Services"
    # Label repair-related activities
    elif "repair" in activity or "maintenance" in activity:
        return "Repair and Maintenance"
    # Default label for anything else
    elif "blank" in activity or "incomplete description/ wrongly classifed" in activity:
        return "NO NIC-NAME"
    else:
        return "Other services"

df_new['Label'] = df_new["NICName"].apply(label_activity)
```

```python
df_new
```

```python
# Step 2: Clustering with KMeans (using TF-IDF for feature extraction)
# Vectorize the activity descriptions
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df_new["NICName"])

# Apply KMeans clustering
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df_new['Cluster'] = kmeans.fit_predict(X)
```

```python
# Display unique combinations of 'Label' and 'Cluster'
unique_label_cluster = df_new[['Label', 'Cluster']].drop_duplicates().sort_values(by='Label')

# Display the result
print(unique_label_cluster)
```
